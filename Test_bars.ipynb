```python
import cv2
import numpy as np
import os
def trim(frame):
    if not np.sum(frame[0]):
        return trim(frame[1:])
    elif not np.sum(frame[-1]):
        return trim(frame[:-2])
    elif not np.sum(frame[:,0]):
        return trim(frame[:,1:]) 
    elif not np.sum(frame[:,-1]):
        return trim(frame[:,:-2])    
    return frame
```

```python
arr = os.listdir('/content/drive/MyDrive/G32V11H0')
for im in arr:
  img = cv2.imread('/content/drive/MyDrive/G32V11H0/'+im)   
  thold = (img>10)*img
  trimmedImage = trim(thold)
  cv2.imwrite('/content/drive/MyDrive/CROP/32v11/'+im,trimmedImage)
```
